```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

def ScaleTrainSplit(x,y):
  scaler=MinMaxScaler()
  x_scale=scaler.fit_transform(x)
  x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.3,random_state=1)
  return scaler,x_train,x_test,y_train,y_test


```

```python
scaler,x_train,x_test,y_train,y_test=ScaleTrainSplit(x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

def KNN_clf(x_train,x_test,y_train,y_test):
  print("KNN "*20)

  knn=KNeighborsClassifier()
  params={'n_neighbors':[3,5,7,9,11,13],
          'weights':['uniform','distance'],
          }
  clf=GridSearchCV(knn,params,cv=10,scoring='accuracy')
  clf.fit(x_train,y_train)
  print(clf.best_params_)
  y_pred=clf.predict(x_test)
  print(classification_report(y_test,y_pred))
  return clf
```

```python
from sklearn.naive_bayes import GaussianNB

def nb(x_train,x_test,y_train,y_test):
  print("NB "*20)
  nb=GaussianNB()
  nb.fit(x_train,y_train)
  y_pred=nb.predict(x_test)
  print(classification_report(y_test,y_pred))
  return nb
```

```python
from sklearn.svm import SVC

def clf_svc(x_train,x_test,y_train,y_test):
  print("SVC "*20)
  svc=SVC()
  params = {
    'kernel':['linear','poly', 'rbf', 'sigmoid'],
    'degree':[1,2,3],
    'gamma':['scale','auto']
}
  clf=GridSearchCV(svc,params,cv=10,scoring='accuracy')
  clf.fit(x_train,y_train)
  print(clf.best_params_)
  y_pred=clf.predict(x_test)
  print(classification_report(y_test,y_pred))
  return clf
```

```python
from sklearn.tree import DecisionTreeClassifier

def clf_dt(x_train,x_test,y_train,y_test):
  print("dt "*20)

  dt=DecisionTreeClassifier()
  params={'criterion':['gini','entropy','log_loss'],
          'splitter':['best','random'],
          'max_depth':[3,4,5]}

  clf=GridSearchCV(dt,params,cv=10,scoring='accuracy')
  clf.fit(x_train,y_train)
  print(clf.best_params_)
  y_pred=clf.predict(x_test)
  print(classification_report(y_test,y_pred))
  return clf
```

```python
from sklearn.ensemble import RandomForestClassifier
def clf_rf(x_train,x_test,y_train,y_test):
  print("rf "*20)

  params={'criterion':['gini','entropy','log_loss'],
            'max_depth':[3,4,5],
          'random_state':[1]}
  rf=RandomForestClassifier()
  clf=GridSearchCV(rf,params,cv=10,scoring='accuracy')
  clf.fit(x_train,y_train)
  print(clf.best_params_)
  y_pred=clf.predict(x_test)
  print(classification_report(y_test,y_pred))
  return clf
```

```python
from sklearn.ensemble import AdaBoostClassifier
def clf_abc(x_train,x_test,y_train,y_test):
  print("abc "*20)
  abc=AdaBoostClassifier()
  abc.fit(x_train,y_train)
  y_pred=abc.predict(x_test)
  print(classification_report(y_test,y_pred))
  return abc
```

```python
from sklearn.ensemble import GradientBoostingClassifier
def clf_gbc(x_train,x_test,y_train,y_test):
  print("gbc "*20)
  gbc=GradientBoostingClassifier()
  gbc.fit(x_train,y_train)
  y_pred=gbc.predict(x_test)
  print(classification_report(y_test,y_pred))
  return gbc
```

```python
from xgboost import XGBClassifier
def clf_xgb(x_train,x_test,y_train,y_test):
  print("xgb "*20)
  xgb=XGBClassifier()
  xgb.fit(x_train,y_train)
  y_pred=xgb.predict(x_test)
  print(classification_report(y_test,y_pred))
  return xgb
```

```python
from sklearn.linear_model import LogisticRegression
def clf_logreg(x_train,x_test,y_train,y_test):
  print("lr "*20)
  lr=LogisticRegression()
  lr.fit(x_train,y_train)
  y_pred=lr.predict(x_test)
  print(classification_report(y_test,y_pred))
  return lr
```
